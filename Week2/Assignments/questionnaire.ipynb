1. What are the main differences between external, internal, and inline CSS?
    
- External CSS refers to a separate document with the styling for the elements in the HTML document.
- Internal CSS refers to the style element found under < head >.
- Inline CSS refers to the style found as an attribute within an element.

7. From inside to outside, what is the order of box-model properties?

9. What is the difference between the standard and alternative box model?

        The standard box model only acounts for the sizing of the content and not the rest of the properties. The alternative box model takes the properties and content sizings into account.

13. What is the difference between a block element and an inline element?

        A block element always starts on a new line and takes up all the width available. An inline element, on the other hand, does not start on a new line and only takes up as much width as necessary.

14. What is the difference between an inline element and an inline-block element?

                An inline element is one which only takes as much width necessary and does not start on a new line. An inline-block element can take block elements and create a single block line with the elements being inline

19. What’s the difference between a flex container and a flex item?

        The flex container acts as a block element to its neighboring elements and contains flex items which are subject to flex parameters and rules.

20. How do you create a flex item?

        By declaring its display property to be flex

21. What are the 3 values defined in the shorthand flex property?

* Flex-Grow: How much available space the flex item will take up.

* Flex-Shrink: When there is not enough space for the items, how their width and height should be affected proportionally to one another.

* Flex-Basis: The minimum value applied to the width and height of a flex item.

22. How do you make flex items arrange themselves vertically instead of horizontally?

        flex-direction: vertical;

23. What is the difference between justify-content and align-items?

        justify-content affects flex items on the main axis (Horizontal axis) and align-items affects flex items on the cross-axis (vertical axis).

24. How do you use flexbox to completely center a div inside a flex container?

        Set both values of justify-content and align-items to center

25. What’s the difference between justify-content: space-between and justify-content: 
space-around?

        space-between distributes the available space between all flex items between them whereas space-around distributes the space evenly between and around the items.
