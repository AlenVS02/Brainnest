CSS Selectors
-

* **Universal Selector** " * " - The asterisk is considered a universal selector. 
    It should be noted that this has the lowest priority of the selectors.
    
* **Type Selector** - Targets elements of a specified type. 

i.e.  

    <div>, <p>, <img>, etc.

* **Class Selector** " . " - selects all element with the same class attribute. Helpful for targeting elements with purpose.

* **ID Selector** " # " - Selects the element with the specified ID. IDs are unique and have the highest priority.
    
i.e.
   
    <nav>
        <div class="navbar-logo" id="cool-element"> Hello Inc. </div>
    </nav>

    .navbar-logo {
            color: red;
    }

* One can group selectors, so that they all share the same styling 

i.e.

    .navbar-logo, div #cool-element {
        color
    }

* **Chained Selector** - When we target multiple selectors. This is done by chaining selectors without any sepparationg.

i.e.

    p.my-id {
        
    }

* **Descendant Combinator** - Nested elements with differing classes. Due to their infrequent use, when applied, their priority is one of the highest. Separated by a space.

Properties
-

Properties are considered the styling rule that we apply to the selected elements.

    propertyy:value

Commonly used properties:

* **Color Property** - Sets the foreground color. (Typically of text)

* **Background-color Property** - Sets the background color of an element.

* **Font-family** - Can be a single valua or a comma-separated list of values that determine what font an element uses. If the browser does not supportthe first font, it will move on to the next one.

* **Font-size** - Determines size of font.

* **Font-weight** - Affects the boldness of text if the fonts suport the specified weight.
    - thin
    - normal
    - thick
    - Numeric Values: 1-800
    <br><br>
* **Text-align** - Will align text horizontally within an element.
    - left
    - right
    - center
    - justify
    - inherit
    <br><br>

* **Image Properties**

    - **Height and Width** - It is common practice to set the height to auto and the width to be adjusted manually. This is done to ensure that when adjusting its size it scales proportionally.
    - **border-radius** - Used to have rounder corners (not exclusive to images.).

Cascade of CSS
-

1. Specificity
2. Inheritance
3. Rule order

**Specificity** - A CSS declaration that is more specific will take precedence over less specific ones. Only comes into play when an element has multiple, conflicting declarations targeting it.

**Inheritance** - Properties applied to a parent element will be inherited by its children element.

**Rule Order** - When there are conflicting declarations (multiple classes), the last one declared will take precedence over the rest.

The Box Model
-

Every element in HTML comes in the form of four-sided boxes, squares, or rectangles.

* **Padding** - How the content will behave (in terms of positioning) inside the box. 
* **Margin** - The position/size of the element relative to other elements.
* **Border** - The area containing the element. It is found between the margin and the padding
